```python
import os
import json
import hashlib
import uuid
from typing import List, Dict, Tuple
from dotenv import load_dotenv
from opensearchpy import OpenSearch, RequestsHttpConnection, AWSV4SignerAuth
import boto3
from twelvelabs import TwelveLabs

load_dotenv()

# Configuration
OPENSEARCH_HOST = os.getenv("OPENSEARCH_HOST")
OPENSEARCH_PORT = int(os.getenv("OPENSEARCH_PORT", 443))
AWS_REGION = os.getenv("AWS_REGION", "us-east-1")
INDEX_NAME = "video_clips_twelvelabs"

# Embedding Model Configuration
USE_BEDROCK_MARENGO = True  # Set to True for Bedrock, False for Twelvelabs
BEDROCK_MODEL_ID = "us.twelvelabs.marengo-embed-2-7-v1:0"
TWELVELABS_MODEL_NAME = "Marengo-retrieval-2.7"

# Initialize clients
client = TwelveLabs(api_key=os.getenv("TWELVELABS_API_KEY"))
s3_client = boto3.client('s3', region_name=AWS_REGION)
bedrock_runtime = boto3.client('bedrock-runtime', region_name=AWS_REGION)


def get_opensearch_client():
    """Initialize OpenSearch client with AWS authentication"""
    opensearch_host = OPENSEARCH_HOST.replace('https://', '').replace('http://', '').strip()
    
    print(f"Connecting to OpenSearch: {opensearch_host}:{OPENSEARCH_PORT}")
    
    session = boto3.Session()
    credentials = session.get_credentials()
    
    if not credentials:
        raise ValueError("AWS credentials not found. Check your AWS configuration.")
    
    # Use AWSV4SignerAuth for both Serverless and Cluster
    auth = AWSV4SignerAuth(credentials, AWS_REGION, 'es')
    
    try:
        opensearch_client = OpenSearch(
            hosts=[{'host': opensearch_host, 'port': OPENSEARCH_PORT}],
            http_auth=auth,
            use_ssl=True,
            verify_certs=True,
            connection_class=RequestsHttpConnection,
            pool_maxsize=20,
            timeout=30
        )
        
        # Test connection
        info = opensearch_client.info()
        print(f" Connected to OpenSearch cluster: {info.get('cluster_name', 'Unknown')}")
        
        return opensearch_client
    
    except Exception as e:
        print(f" OpenSearch Connection Error: {type(e).__name__}")
        print(f"   Message: {str(e)}")
        print(f"\n   Troubleshooting:")
        print(f"   1. Verify OPENSEARCH_HOST is correct: {opensearch_host}")
        print(f"   2. Verify OPENSEARCH_PORT is correct: {OPENSEARCH_PORT}")
        print(f"   3. Check AWS credentials are valid")
        print(f"   4. Ensure security group allows access")
        print(f"   5. Verify IAM permissions for 'es:*' actions")
        raise


def generate_bedrock_marengo_embedding(text: str) -> List[float]:
    """Generate embedding using Bedrock Marengo model"""
    try:
        print(f"    Generating embedding using Bedrock Marengo...")
        
        request_body = {
            "inputType": "text",
            "inputText": text,
            "textTruncate": "none"
        }
        
        response = bedrock_runtime.invoke_model(
            modelId=BEDROCK_MODEL_ID,
            body=json.dumps(request_body),
            contentType="application/json",
            accept="application/json"
        )
        
        result = json.loads(response['body'].read())
        
        if 'data' in result and len(result['data']) > 0:
            embedding = result['data'][0].get('embedding', [])
            print(f"    Generated Bedrock embedding with {len(embedding)} dimensions")
            return embedding
        else:
            print(f"    No embedding found in Bedrock response")
            return []
        
    except Exception as e:
        print(f"    Error generating Bedrock embedding: {e}")
        return []


def generate_twelvelabs_marengo_embedding(text: str) -> List[float]:
    """Generate embedding using Twelvelabs Marengo model"""
    try:
        print(f"    Generating embedding using Twelvelabs Marengo...")
        
        # Create a task to generate embedding for the query text
        task_result = client.embed.tasks.create(
            model_name=TWELVELABS_MODEL_NAME,
            text=text
        )
        
        if task_result.text_embedding and task_result.text_embedding.segments:
            embedding = task_result.text_embedding.segments[0].float_
            print(f"    Generated Twelvelabs embedding with {len(embedding)} dimensions")
            return embedding
        else:
            print(f"    No embedding found in Twelvelabs task result")
            return None
    
    except Exception as e:
        print(f"    Error generating Twelvelabs embedding: {e}")
        return None


def generate_marengo_embedding(text: str, use_bedrock: bool = None) -> List[float]:
    """Generate embedding using configured Marengo model (Bedrock or Twelvelabs)"""
    # Use global flag if not specified
    bedrock_flag = use_bedrock if use_bedrock is not None else USE_BEDROCK_MARENGO
    
    if bedrock_flag:
        return generate_bedrock_marengo_embedding(text)
    # else:
    #     return generate_twelvelabs_marengo_embedding(text)


def create_index_if_not_exists(opensearch_client):
    """Create video_clips index if it doesn't exist"""
    try:
        if not opensearch_client.indices.exists(index=INDEX_NAME):
            index_body = {
                "settings": {
                    "index": {
                        "knn": True,
                        "number_of_shards": 2,
                        "number_of_replicas": 1
                    }
                },
                "mappings": {
                    "properties": {
                        "video_id": {"type": "keyword"},
                        "video_title": {"type": "text"},
                        "clip_id": {"type": "keyword"},
                        "segment_index": {"type": "integer"},
                        "timestamp_start": {"type": "float"},
                        "timestamp_end": {"type": "float"},
                        "embedding_scope": {"type": "keyword"},  # "clip"
                        "embedding_option": {"type": "keyword"},  # "visual-text", "audio"
                        "embedding": {
                            "type": "knn_vector",
                            "dimension": 1024,
                            "method": {
                                "name": "hnsw",
                                "space_type": "l2",
                                "engine": "lucene",
                                "parameters": {
                                    "ef_construction": 128,
                                    "m": 16
                                }
                            }
                        }
                    }
                }
            }
            
            opensearch_client.indices.create(index=INDEX_NAME, body=index_body)
            print(f" Created index: {INDEX_NAME}")
        else:
            print(f" Index {INDEX_NAME} already exists")
    
    except Exception as e:
        print(f" Error creating index: {type(e).__name__}")
        print(f"   Message: {str(e)}")
        if "403" in str(e) or "Forbidden" in str(e):
            print(f"\n   This is a permissions error. Ensure your IAM user has:")
            print(f"   - es:CreateIndex")
            print(f"   - es:PutMapping")
            print(f"   - es:DescribeElasticsearchDomain")
        raise


def generate_clip_id(video_id: str, start_time: float, end_time: float) -> str:
    """Generate deterministic clip_id based on video_id and timestamps"""
    clip_string = f"{video_id}_{start_time:.2f}_{end_time:.2f}"
    clip_hash = hashlib.sha256(clip_string.encode()).hexdigest()[:16]
    return f"clip_{clip_hash}"


def validate_embedding(embedding, expected_dim=1024) -> Tuple[bool, str]:
    """Validate embedding dimensions and data"""
    if not isinstance(embedding, list):
        return False, f"Embedding is not a list: {type(embedding)}"
    
    if len(embedding) != expected_dim:
        return False, f"Embedding dimension mismatch: expected {expected_dim}, got {len(embedding)}"
    
    for i, val in enumerate(embedding):
        if not isinstance(val, (int, float)):
            return False, f"Embedding contains non-numeric value at index {i}: {val}"
        if val != val:  # Check for NaN
            return False, f"Embedding contains NaN at index {i}"
    
    return True, "Valid"


def index_video_embeddings(opensearch_client, video_id: str, video_title: str, 
                           segments: List[Dict]) -> int:
    """Index embeddings from a single video into OpenSearch"""
    indexed_count = 0
    skipped_count = 0
    
    print(f"\n Indexing video: {video_title} ({video_id})")
    print(f"   Total segments: {len(segments)}")
    
    for idx, segment in enumerate(segments):
        try:
            # Extract embedding
            embedding = segment.float_
            
            # Validate embedding
            is_valid, validation_msg = validate_embedding(embedding)
            if not is_valid:
                print(f"    Skipping segment {idx}: {validation_msg}")
                skipped_count += 1
                continue
            
            # Get timestamps
            start_time = segment.start_offset_sec
            end_time = segment.end_offset_sec
            
            # Generate deterministic clip_id
            clip_id = generate_clip_id(video_id, start_time, end_time)
            
            # Prepare document
            doc = {
                'video_id': video_id,
                'video_title': video_title,
                'clip_id': clip_id,
                'segment_index': idx,
                'timestamp_start': float(start_time),
                'timestamp_end': float(end_time),
                'embedding_scope': segment.embedding_scope,
                'embedding_option': segment.embedding_option,
                'embedding': embedding
            }
            
            # Index document
            response = opensearch_client.index(
                index=INDEX_NAME,
                body=doc
            )
            
            indexed_count += 1
            
            if indexed_count % 5 == 0:
                print(f"    Indexed {indexed_count} segments...")
        
        except Exception as e:
            print(f"    Error indexing segment {idx}: {e}")
            skipped_count += 1
            continue
    
    print(f"    Successfully indexed {indexed_count} segments (skipped: {skipped_count})")
    return indexed_count


def index_all_videos_from_twelvelabs(index_id: str, video_ids: List[str], 
                                     video_titles: Dict[str, str] = None) -> Dict:
    """Index all videos from Twelvelabs into OpenSearch"""
    opensearch_client = get_opensearch_client()
    create_index_if_not_exists(opensearch_client)
    
    if video_titles is None:
        video_titles = {}
    
    total_indexed = 0
    results = {
        'total_videos': len(video_ids),
        'videos_processed': 0,
        'total_segments_indexed': 0,
        'errors': []
    }
    
    print(f"\n Starting indexing of {len(video_ids)} videos from Twelvelabs...")
    print(f"   Index ID: {index_id}")
    print(f"   OpenSearch Index: {INDEX_NAME}\n")
    
    for video_id in video_ids:
        try:
            # Retrieve video with embeddings
            print(f"\n Retrieving embeddings for video: {video_id}")
            video = client.indexes.videos.retrieve(
                index_id=index_id,
                video_id=video_id,
                embedding_option=["visual-text", "audio"]
            )
            
            # Get video title
            video_title = video_titles.get(video_id, video.title or video_id)
            
            # Extract segments
            if video.embedding and video.embedding.video_embedding and video.embedding.video_embedding.segments:
                segments = video.embedding.video_embedding.segments
                
                # Index embeddings
                indexed = index_video_embeddings(
                    opensearch_client,
                    video_id,
                    video_title,
                    segments
                )
                
                total_indexed += indexed
                results['videos_processed'] += 1
                results['total_segments_indexed'] += indexed
            else:
                print(f"    No embeddings found for video {video_id}")
                results['errors'].append(f"No embeddings for {video_id}")
        
        except Exception as e:
            error_msg = f"Error processing video {video_id}: {str(e)}"
            print(f"    {error_msg}")
            results['errors'].append(error_msg)
    
    print(f"\n" + "="*60)
    print(f" INDEXING COMPLETE")
    print(f"="*60)
    print(f"Videos processed: {results['videos_processed']}/{results['total_videos']}")
    print(f"Total segments indexed: {results['total_segments_indexed']}")
    if results['errors']:
        print(f"Errors: {len(results['errors'])}")
        for error in results['errors']:
            print(f"  - {error}")
    print(f"="*60 + "\n")
    
    return results


# Configuration Display
print("=" * 70)
print(" EMBEDDING MODEL CONFIGURATION")
print("=" * 70)
print(f"  Embedding Model: {'Bedrock Marengo' if USE_BEDROCK_MARENGO else 'Twelvelabs Marengo'}")
if USE_BEDROCK_MARENGO:
    print(f"  Bedrock Model ID: {BEDROCK_MODEL_ID}")
else:
    print(f"  Twelvelabs Model: {TWELVELABS_MODEL_NAME}")
print(f"  OpenSearch Host: {OPENSEARCH_HOST}")
print(f"  OpenSearch Port: {OPENSEARCH_PORT}")
print(f"  AWS Region: {AWS_REGION}")
print(f"  Index Name: {INDEX_NAME}")
print("=" * 70)
```

```python
# Fetch records from OpenSearch index using Marengo embeddings (Bedrock or Twelvelabs)
import json

print("=" * 70)
print(" FETCH RECORDS FROM OPENSEARCH USING MARENGO EMBEDDINGS")
print("=" * 70)


def search_opensearch_with_marengo(opensearch_client, query_text: str, 
                                   top_k: int = 10, use_bedrock: bool = None) -> Dict:
    """Search OpenSearch using Marengo-generated embedding (Bedrock or Twelvelabs)"""
    try:
        print(f"\n Searching for: '{query_text}'")
        print(f"   Top K: {top_k}")
        
        # Generate embedding using configured model
        query_embedding = generate_marengo_embedding(query_text, use_bedrock=use_bedrock)
        
        if not query_embedding:
            return {
                'query': query_text,
                'success': False,
                'error': 'Failed to generate query embedding',
                'results': []
            }
        
        # Search OpenSearch using KNN
        search_body = {
            "size": top_k,
            "query": {
                "knn": {
                    "embedding": {
                        "vector": query_embedding,
                        "k": top_k
                    }
                }
            }
        }
        
        print(f"\n    Searching OpenSearch index: {INDEX_NAME}")
        response = opensearch_client.search(index=INDEX_NAME, body=search_body)
        
        # Process results
        results = []
        for hit in response['hits']['hits']:
            doc = hit['_source']
            results.append({
                'score': float(hit['_score']),
                'video_id': doc.get('video_id'),
                'video_title': doc.get('video_title'),
                'clip_id': doc.get('clip_id'),
                'segment_index': doc.get('segment_index'),
                'timestamp_start': doc.get('timestamp_start'),
                'timestamp_end': doc.get('timestamp_end'),
                'embedding_scope': doc.get('embedding_scope'),
                'embedding_option': doc.get('embedding_option'),
                'embedding_dimension': len(doc.get('embedding', []))
            })
        
        print(f"    Found {len(results)} results")
        
        return {
            'query': query_text,
            'success': True,
            'total_results': len(results),
            'top_k': top_k,
            'embedding_model': 'Bedrock Marengo' if (use_bedrock if use_bedrock is not None else USE_BEDROCK_MARENGO) else 'Twelvelabs Marengo',
            'results': results
        }
    
    except Exception as e:
        print(f"    Search error: {e}")
        return {
            'query': query_text,
            'success': False,
            'error': str(e),
            'results': []
        }


def fetch_all_indexed_records(opensearch_client, limit: int = 100) -> Dict:
    """Fetch all records from the index"""
    try:
        print(f"\n Fetching all indexed records (limit: {limit})")
        
        search_body = {
            "size": limit,
            "query": {"match_all": {}},
            "_source": [
                "video_id", "video_title", "clip_id", "segment_index",
                "timestamp_start", "timestamp_end", "embedding_scope",
                "embedding_option"
            ]
        }
        
        response = opensearch_client.search(index=INDEX_NAME, body=search_body)
        
        records = []
        for hit in response['hits']['hits']:
            doc = hit['_source']
            records.append({
                'id': hit['_id'],
                'video_id': doc.get('video_id'),
                'video_title': doc.get('video_title'),
                'clip_id': doc.get('clip_id'),
                'segment_index': doc.get('segment_index'),
                'timestamp_start': doc.get('timestamp_start'),
                'timestamp_end': doc.get('timestamp_end'),
                'embedding_scope': doc.get('embedding_scope'),
                'embedding_option': doc.get('embedding_option')
            })
        
        total_count = response['hits']['total']['value']
        
        print(f"    Retrieved {len(records)} records (total in index: {total_count})")
        
        return {
            'success': True,
            'total_in_index': total_count,
            'records_fetched': len(records),
            'records': records
        }
    
    except Exception as e:
        print(f"    Error fetching records: {e}")
        return {
            'success': False,
            'error': str(e),
            'records': []
        }


# Test 1: Fetch all indexed records
print("\n\n" + "=" * 70)
print("TEST 1: FETCH ALL INDEXED RECORDS")
print("=" * 70)

try:
    opensearch_client = get_opensearch_client()    
except Exception as e:
    print(f" Error: {e}")


# Test 2: Search using Marengo embeddings (Bedrock or Twelvelabs)
print("\n\n" + "=" * 70)
print("TEST 2: SEARCH USING MARENGO EMBEDDINGS")
print("=" * 70)

# Define search queries
search_queries = [
    "mountain dew"
]

try:
    opensearch_client = get_opensearch_client()
    
    # Test with current model (USE_BEDROCK_MARENGO flag)
    print(f"\n{'='*70}")
    print(f" Using: {'Bedrock Marengo' if USE_BEDROCK_MARENGO else 'Twelvelabs Marengo'}")
    print(f"{'='*70}")
    
    all_search_results = {
        'index': INDEX_NAME,
        'embedding_model': 'Bedrock Marengo' if USE_BEDROCK_MARENGO else 'Twelvelabs Marengo',
        'searches': []
    }
    
    for query in search_queries:
        print(f"\n{'='*70}")
        search_result = search_opensearch_with_marengo(
            opensearch_client,
            query_text=query,
            top_k=5
        )
        
        all_search_results['searches'].append(search_result)
        
        # Print individual search result as JSON
        print(f"\n Search Result for '{query}' (JSON):")
        print(json.dumps(search_result, indent=2))    
except Exception as e:
    print(f" Error: {e}")
    import traceback
    traceback.print_exc()
```

```python
# Update OpenSearch Index with Video Paths
import json
from datetime import datetime
from typing import Dict

print("=" * 70)
print(" UPDATE OPENSEARCH INDEX WITH VIDEO PATHS")
print("=" * 70)

# Define video ID to S3 path mappings
VIDEO_PATH_MAPPINGS = {
    "690c7bd4952dd95fc6e2aa8f": "s3://condenast-raw-useast1-943143228843-dev/year=2025/month=11/day=05/Mountain_Dew_with_Charlie_Day_Zach_Lavine.mp4",
    "690c7be5952dd95fc6e2aaa9": "s3://condenast-raw-useast1-943143228843-dev/year=2025/month=10/day=31/Jurassic-World-Rebirth-_-Official-Trailer-2.mp4",
    "690c7be3ce637f9e1ec3d97d": "s3://condenast-raw-useast1-943143228843-dev/year=2025/month=11/day=05/The_Dancer_Director_s_Cut.mp4",
    "690c7be2952dd95fc6e2aaa6": "s3://condenast-raw-useast1-943143228843-dev/year=2025/month=11/day=05/Apple_iPhone_-Ad.mp4",
    "690c7bde3d0db4ad0948d93d": "s3://condenast-raw-useast1-943143228843-dev/year=2025/month=11/day=05/Gordon-Ramsay_In_Kitchen_MasterClass.mp4"
}


def get_opensearch_client():
    """Initialize OpenSearch client with AWS authentication"""
    opensearch_host = OPENSEARCH_HOST.replace('https://', '').replace('http://', '').strip()
    
    print(f"Connecting to OpenSearch: {opensearch_host}:{OPENSEARCH_PORT}")
    
    session = boto3.Session()
    credentials = session.get_credentials()
    
    if not credentials:
        raise ValueError("AWS credentials not found. Check your AWS configuration.")
    
    # Use AWSV4SignerAuth for both Serverless and Cluster
    auth = AWSV4SignerAuth(credentials, AWS_REGION, 'es')
    
    try:
        opensearch_client = OpenSearch(
            hosts=[{'host': opensearch_host, 'port': OPENSEARCH_PORT}],
            http_auth=auth,
            use_ssl=True,
            verify_certs=True,
            connection_class=RequestsHttpConnection,
            pool_maxsize=20,
            timeout=30
        )
        
        # Test connection
        info = opensearch_client.info()
        print(f" Connected to OpenSearch cluster: {info.get('cluster_name', 'Unknown')}")
        
        return opensearch_client
    
    except Exception as e:
        print(f" OpenSearch Connection Error: {type(e).__name__}")
        print(f"   Message: {str(e)}")
        print(f"\n   Troubleshooting:")
        print(f"   1. Verify OPENSEARCH_HOST is correct: {opensearch_host}")
        print(f"   2. Verify OPENSEARCH_PORT is correct: {OPENSEARCH_PORT}")
        print(f"   3. Check AWS credentials are valid")
        print(f"   4. Ensure security group allows access")
        print(f"   5. Verify IAM permissions for 'es:*' actions")
        raise


# Initialize OpenSearch client for this cell
print("\n Initializing OpenSearch client...")
try:
    opensearch_client_update = get_opensearch_client()
    print(" OpenSearch client initialized successfully")
except Exception as e:
    print(f" Error initializing OpenSearch client: {e}")
    opensearch_client_update = None


def update_video_path_by_video_id(opensearch_client, video_id: str, video_path: str) -> Dict:
    """Update video_path for all documents with given video_id"""
    try:
        print(f"\n Updating video_path for video_id: {video_id}")
        
        # Query to find all documents with this video_id
        query_body = {
            "query": {
                "term": {
                    "video_id": video_id
                }
            }
        }
        
        # Search for documents
        search_response = opensearch_client.search(
            index=INDEX_NAME,
            body=query_body,
            size=1000  # Get all matching documents
        )
        
        total_hits = search_response['hits']['total']['value']
        print(f"   Found {total_hits} documents with video_id: {video_id}")
        
        # Update each document
        updated_count = 0
        errors = []
        
        for hit in search_response['hits']['hits']:
            doc_id = hit['_id']
            
            try:
                # Update document with new video_path
                update_body = {
                    "doc": {
                        "video_path": video_path
                    }
                }
                
                update_response = opensearch_client.update(
                    index=INDEX_NAME,
                    id=doc_id,
                    body=update_body
                )
                
                updated_count += 1
                
            except Exception as e:
                error_msg = f"Error updating doc {doc_id}: {str(e)}"
                errors.append(error_msg)
                print(f"    {error_msg}")
        
        print(f"    Successfully updated {updated_count} documents")
        
        return {
            'video_id': video_id,
            'video_path': video_path,
            'total_found': total_hits,
            'updated': updated_count,
            'errors': errors,
            'success': len(errors) == 0
        }
    
    except Exception as e:
        error_msg = f"Error updating video_id {video_id}: {str(e)}"
        print(f"    {error_msg}")
        return {
            'video_id': video_id,
            'video_path': video_path,
            'success': False,
            'error': error_msg
        }


def batch_update_video_paths(opensearch_client, video_path_mappings):
    """Batch update video paths for multiple videos"""
    print(f"\n Starting batch update of {len(video_path_mappings)} videos...")
    
    results = {
        'timestamp': datetime.now().isoformat(),
        'index': INDEX_NAME,
        'total_videos': len(video_path_mappings),
        'updates': [],
        'summary': {
            'successful': 0,
            'failed': 0,
            'total_documents_updated': 0
        }
    }
    
    for video_id, video_path in video_path_mappings.items():
        update_result = update_video_path_by_video_id(
            opensearch_client,
            video_id,
            video_path
        )
        
        results['updates'].append(update_result)
        
        if update_result['success']:
            results['summary']['successful'] += 1
            results['summary']['total_documents_updated'] += update_result.get('updated', 0)
        else:
            results['summary']['failed'] += 1
    
    print(f"\n" + "=" * 70)
    print(f" BATCH UPDATE COMPLETE")
    print(f"=" * 70)
    print(f"Successful: {results['summary']['successful']}/{results['total_videos']}")
    print(f"Failed: {results['summary']['failed']}/{results['total_videos']}")
    print(f"Total documents updated: {results['summary']['total_documents_updated']}")
    print(f"=" * 70 + "\n")
    
    return results


def generate_opensearch_dev_tools_requests(video_path_mappings):
    """Generate OpenSearch Dev Tools API requests for manual updates"""
    requests = []
    
    for video_id, video_path in video_path_mappings.items():
        # Generate update by query request
        request = {
            "method": "POST",
            "endpoint": f"/{INDEX_NAME}/_update_by_query",
            "body": {
                "query": {
                    "term": {
                        "video_id": video_id
                    }
                },
                "script": {
                    "source": "ctx._source.video_path = params.video_path",
                    "params": {
                        "video_path": video_path
                    }
                }
            }
        }
        requests.append(request)
    
    return requests


# Execute batch update
print("\n" + "=" * 70)
print("STEP 1: BATCH UPDATE VIDEO PATHS")
print("=" * 70)

batch_results = None
if opensearch_client_update:
    try:
        batch_results = batch_update_video_paths(
            opensearch_client_update,
            VIDEO_PATH_MAPPINGS
        )
        
        print("\n Batch Update Results (JSON):")
        print(json.dumps(batch_results, indent=2))
        
    except Exception as e:
        print(f" Error during batch update: {e}")
        import traceback
        traceback.print_exc()
else:
    print(" OpenSearch client not initialized. Skipping batch update.")


# Generate Dev Tools requests
print("\n" + "=" * 70)
print("STEP 2: OPENSEARCH DEV TOOLS API REQUESTS")
print("=" * 70)

dev_tools_requests = generate_opensearch_dev_tools_requests(VIDEO_PATH_MAPPINGS)

print("\n Dev Tools Requests (use in OpenSearch Kibana Console):\n")

for idx, request in enumerate(dev_tools_requests, 1):
    print(f"--- Request {idx} ---")
    print(f"POST {request['endpoint']}")
    print(json.dumps(request['body'], indent=2))
    print()

# Generate combined request
print("\n" + "=" * 70)
print("STEP 3: COMBINED UPDATE BY QUERY REQUEST")
print("=" * 70)

combined_request = {
    "query": {
        "bool": {
            "should": [
                {"term": {"video_id": video_id}}
                for video_id in VIDEO_PATH_MAPPINGS.keys()
            ]
        }
    },
    "script": {
        "source": """
        if (params.mappings.containsKey(doc['video_id'].value)) {
            ctx._source.video_path = params.mappings[doc['video_id'].value];
        }
        """,
        "params": {
            "mappings": VIDEO_PATH_MAPPINGS
        }
    }
}

print(f"\nPOST /{INDEX_NAME}/_update_by_query")
print(json.dumps(combined_request, indent=2))

# Verify updates
print("\n" + "=" * 70)
print("STEP 4: VERIFY UPDATES")
print("=" * 70)

verification_results = None
if opensearch_client_update:
    try:
        verification_results = {
            'timestamp': datetime.now().isoformat(),
            'verified_videos': []
        }
        
        for video_id in VIDEO_PATH_MAPPINGS.keys():
            # Query to get sample document
            query_body = {
                "query": {
                    "term": {
                        "video_id": video_id
                    }
                },
                "size": 1
            }
            
            response = opensearch_client_update.search(index=INDEX_NAME, body=query_body)
            
            if response['hits']['hits']:
                doc = response['hits']['hits'][0]['_source']
                verification_results['verified_videos'].append({
                    'video_id': video_id,
                    'video_path': doc.get('video_path', 'NOT SET'),
                    'expected_path': VIDEO_PATH_MAPPINGS[video_id],
                    'verified': doc.get('video_path') == VIDEO_PATH_MAPPINGS[video_id]
                })
        
        print("\n Verification Results (JSON):")
        print(json.dumps(verification_results, indent=2))
        
    except Exception as e:
        print(f" Error during verification: {e}")
else:
    print(" OpenSearch client not initialized. Skipping verification.")


# Export all updates to JSON
print("\n" + "=" * 70)
print("STEP 5: EXPORT RESULTS")
print("=" * 70)

export_data = {
    'timestamp': datetime.now().isoformat(),
    'index': INDEX_NAME,
    'batch_update_results': batch_results,
    'verification_results': verification_results,
    'dev_tools_requests': dev_tools_requests
}

print("\n Complete Results (JSON):")
print(json.dumps(export_data, indent=2))
```
